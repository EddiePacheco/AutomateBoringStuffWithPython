**The shutil module**

```python
from pathlib import Path
import shutil

HomePath = Path.home()
hello_file = HomePath / 'hello.txt'
some_folder = HomePath / 'some_folder'

# Create 'hello.txt' if it doesn't exist
if not hello_file.exists():
    hello_file.write_text('Hello, world!')

# Ensure 'some_folder is a directory
if some_folder.exists() and not some_folder.is_dir():
    raise NotADirectoryError(f'{some_folder} is not a directory')

# Create 'some_folder' if it doesn't exist
some_folder.mkdir(exist_ok=True)

# Copy 'hello.txt' to 'some_folder'
shutil.copy(hello_file, some_folder / 'hello.txt')
```

```python
TestFile = HomePath /  'test_file.txt'
if not TestFile.exists():
    TestFile.write_text('Copy pasta!')

shutil.copytree(HomePath / 'copy_test', HomePath / 'paste_test', dirs_exist_ok=True)
```

**Safe deletes with send2trash**

```python
import send2trash

with open('bacon.txt', 'a') as BaconFile:
    BaconFile.write('Bacon is not a vegetable.')

send2trash.send2trash('bacon.txt')
```

**Walking a directory tree**

```python
import os

for Folders, SubFolders, FileNames in os.walk(Path.cwd()):
    print(f'Current folder is: {Folders}')
    for SubFolder in SubFolders:
        print(f'Subfolders inside {Folders}: {SubFolder}')
    for Filename in FileNames:
        print(f'Files inside {Folders}: {Filename}')
    print('')
```
