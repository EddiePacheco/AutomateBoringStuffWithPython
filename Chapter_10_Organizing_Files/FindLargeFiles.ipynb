```python
# Find large files in a directory tree

# This script walks through a directory tree and prints the names of files that
# are larger than a certain size. The default size is 100MB.

# Step 1: Import modules
# Step 2: Walk through the directory tree
# Step 3: Print the names of large files
```

```python
import os
from pathlib import Path
```

```python
for foldername, subfolders, filenames in os.walk('.'):
    for filename in filenames:
        file_path = Path(foldername) / filename
        try:
            if file_path.is_file() and os.path.getsize(file_path) > 100000000:
                print(f'File: {file_path}')
                print(f'Size: {os.path.getsize(file_path)} bytes')
        except PermissionError:
            print(f'Permission denied: {file_path}')
```
