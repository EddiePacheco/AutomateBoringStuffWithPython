**Saving variables with the pprint.pformat() function**

```python
import pprint
```

```python
cats = [{'name': 'Zophie', 'desc': 'chubby'}, {'name': 'Pooka', 'desc': 'fluffy'}]
pprint.pformat(cats)
```

```python
FileObj = open('MyCats.py', 'w')
FileObj.write('cats = ' + pprint.pformat(cats) + '\n')
```
